```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

#above we have filled the nulls in categorical data with their respective mode. 
#Below we will fill the nulls with mean value in numeric data.

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].mean())
```

```python
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
```

```python
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].map({'Male':1, 'Female':0})
df['Married'] = df['Married'].map({'Yes':1, 'No':0})
df['Education'] = df['Education'].map({'Graduate':1, 'Not Graduate':0})
df['Self_Employed'] = df['Self_Employed'].map({'Yes':1, 'No':0})
df['Property_Area'] = df['Property_Area'].map({'Urban':2, 'Semiurban':1, 'Rural':0})
df['Loan_Status']= df['Loan_Status'].map({'Y':1, 'N':0})
```

```python
df.head()
```

```python
df = df.drop('Loan_ID', axis=1)
```

```python
df.head()
```

```python
df['Dependents'] = df['Dependents'].replace('3+', 3)
df['Dependents'] = df['Dependents'].astype(int)
```

```python
print(df.dtypes)
```

#now we will create ML model, check accuracy, confusion matrix

```python
#define Features(X), and Target(Y)
X = df.drop('Loan_Status', axis=1)
Y = df['Loan_Status']
```

```python
#train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#import model
from sklearn.linear_model import LogisticRegression
```

```python
#train model
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
#cheching the unique values of Dependents
print(df['Dependents'].unique())
```

```python
#model evaluation; checking accuracy

from sklearn.metrics import accuracy_score

Y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)

print("Accuracy:", accuracy)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test, Y_pred))
```

#Model Predicts approvals well but struggles to correctly identify risky customers.
#so, improvements is needed in detecting risky customers to reduce false approvals.

```python
#trying custom prediction
new_data = [[1, 1, 0, 1, 0, 5000, 2000, 150, 360, 1, 2]]    #example
prediction = model.predict(new_data)
print(prediction)
```
